```python
plt.figure(figsize=(15,5))
sns.countplot(y='Gender',data=data)
plt.show()
```

```python
age18_30 = data['Age'][(data['Age'] >= 18) & (data['Age'] <= 30)]
age31_40 = data['Age'][(data['Age'] >= 31) & (data['Age'] <= 40)]
age41_50 = data['Age'][(data['Age'] >= 41) & (data['Age'] <= 50)]
age51_60 = data['Age'][(data['Age'] >= 51) & (data['Age'] <= 60)]

x = ['18-30', '31-40', '41-50', '51-60']
y = [len(age18_30.values), len(age31_40.values), len(age41_50.values), len(age51_60.values)]

plt.figure(figsize=(20, 5))
sns.barplot(x=x, y=y, palette='Spectral')  
plt.title('Age')
plt.xlabel('Age')
plt.ylabel('Number of customers')
plt.show()
```

```python
sal40k_50k = data['EstimatedSalary'][(data['EstimatedSalary'] >= 40000) & (data['EstimatedSalary'] <= 50000)]
sal50k_75k = data['EstimatedSalary'][(data['EstimatedSalary'] >= 50001) & (data['EstimatedSalary'] <= 750000)]
sal75k_100k = data['EstimatedSalary'][(data['EstimatedSalary'] >= 75001) & (data['EstimatedSalary'] <= 100000)]
sal100k_125k = data['EstimatedSalary'][(data['EstimatedSalary'] >= 100001) & (data['EstimatedSalary'] <= 125000)]
sal125k_150k = data['EstimatedSalary'][(data['EstimatedSalary'] >= 125001) & (data['EstimatedSalary'] <= 150000)]
x = ['40k-50k', '50k-75k', '75k-100k', '100k-125k','125k-150k']
y = [len(sal40k_50k),len(sal50k_75k),len(sal75k_100k),len(sal100k_125k),len(sal125k_150k)]

plt.figure(figsize=(20, 5))
sns.barplot(x=x, y=y, palette='Spectral')  
plt.title('Estimated salary')
plt.xlabel('salary')
plt.ylabel('Number of customers')
plt.show()
```

```python
x1 = data.loc[:,['Age','EstimatedSalary']].values
from sklearn.cluster import KMeans
k_values = range(1, 11) 

n = []

# Run K-means for each k and store the inertia
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x1)
    n.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(k_values, n, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distances (Inertia)')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
n= kmeans.fit_predict(x1)
n
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x1[:,0],x1[:,1],c = kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.title("Clusters")
plt.xlabel('Age')
plt.ylabel('Estimated salary')
plt.show()
```
